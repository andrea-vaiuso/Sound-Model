```python
ref_manoeuvre = "Hover"
ref_zeta = 0
C_proc_yaw = -3.8256493
```

```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv('Hover_Database.csv', sep=';')
df2 = pd.read_csv('Yaw_Database.csv', sep=';')

# Merge the dataframes
merged_df = pd.concat([df1, df2], ignore_index=True)
# Display the merged dataframe
merged_df.head()
```

```python
# Filter the dataframe
filtered_df = merged_df[(merged_df['zeta[rad]'] == ref_zeta) & (merged_df['manoeuvre_name'] == ref_manoeuvre)]

# Extract the "Lws[dB]" column
lws_values = filtered_df['Lws[dB]'].apply(eval).tolist()

print(lws_values)
print(len(lws_values), ",", len(lws_values[0]))
```

```python
import numpy as np

# Convert lws_values to a numpy array
lws_array = np.array(lws_values)

# Calculate the average by column
average_lws = np.mean(lws_array, axis=0)

# Print the result vector
print(average_lws)
print(len(average_lws))
```

```python
# Filter the dataframe for the specified conditions
hover_zeta_zero_df = merged_df[(merged_df['manoeuvre_name'] == ref_manoeuvre) & (merged_df['zeta[rad]'] == ref_zeta)]

# Calculate the average values for RPM1, RPM2, RPM3, and RPM4
average_rpm1 = hover_zeta_zero_df['RPM1'].mean()
average_rpm2 = hover_zeta_zero_df['RPM2'].mean()
average_rpm3 = hover_zeta_zero_df['RPM3'].mean()
average_rpm4 = hover_zeta_zero_df['RPM4'].mean()

# Print the results
print(f"Average RPM1: {average_rpm1}")
print(f"Average RPM2: {average_rpm2}")
print(f"Average RPM3: {average_rpm3}")
print(f"Average RPM4: {average_rpm4}")
```

```python
# Create new columns by subtracting the mean values
merged_df['delta_RPM1'] = merged_df['RPM1'] - average_rpm1
merged_df['delta_RPM2'] = merged_df['RPM2'] - average_rpm2
merged_df['delta_RPM3'] = merged_df['RPM3'] - average_rpm3
merged_df['delta_RPM4'] = merged_df['RPM4'] - average_rpm4
merged_df['delta_zeta'] = merged_df['zeta[rad]']

# Add a new column "C_proc" based on the manoeuvre_name
merged_df['C_proc'] = merged_df['manoeuvre_name'].apply(lambda x: 0 if x == 'Hover' else C_proc_yaw if x == 'Yaw' else None)

# Display the updated dataframe
merged_df.head()
```

```python
# Convert the average_lws array to a string format
average_lws_str = str(list(average_lws))

# Add the Lw_Ref column to the merged_df dataframe
merged_df['Lw_Ref'] = average_lws_str

# Display the updated dataframe
merged_df.head()
```

```python
merged_df.to_csv('final_dataset.csv', index=False)
```
